# Correlation lists statistics
This notebook contains statistics across different versions of the manual curation list. With it we can see trends, measure what we improved and identify sudden outliers that can indicate manually introduced errors.

```python
import pandas as pd
import sys, os

sys.path.insert(1, os.pardir)

import utils_stats
import utils_plot
```

## Person correlation list

```python
# The correlation lists from 2023-02-03 and 2023-05-05 have a different structure and hence are excluded
# There is also no correlation list for 2023-11-14
personCorrelationLists = {
    '2023-09-01': '2023-09-01_person_contributors-correlation-list.csv',
    '2023-10-09': '2023-10-09_person_contributors-correlation-list.csv',
    '2023-11-08': '2023-11-08_person_contributors-correlation-list.csv',
    '2023-12-06': '2023-12-06_person_contributors-correlation-list.csv',
    '2023-12-15': '2023-12-15_person_contributors-correlation-list.csv',
    '2024-02-05': '2024-02-05_person_contributors-correlation-list.csv',
    '2024-03-27': '2024-03-27_person_contributors-correlation-list.csv',
    '2024-04-23': '2024-04-23_person_contributors-correlation-list.csv',
    '2024-06-12': '2024-06-12_person_contributors-correlation-list.csv',
    '2025-02-10': '2025-02-10_person_contributors-correlation-list.csv'
}
```

```python
# Recalculate the measurements defined in utils_stats and store measurements as CSV files
utils_stats.redoCorrelationListPersonsMeasurements(
    personCorrelationLists, 
    '../../data-sources/correlation'
)
```

```python
personCorrelations = utils_stats.mergeMeasurementsToDataFrame('./measurements/correlation', personCorrelationLists.values())
```

---
### Data source identifiers

```python
sourceIdentifierColumns = ['withKBRIdentifier', 'withBnFIdentifier', 
                                   'withNTAIdentifier', 'withUnescoIdentifier',
                                   'withISNIIdentifier', 'withVIAFIdentifier']
utils_plot.plotPercentageTimeline(personCorrelations,
                                  sourceIdentifierColumns,
                                  'Person correlation list - identifiers',
                                 'numberEntries')
```

```python
personCorrelations[['numberEntries'] + sourceIdentifierColumns]
```

---
### Additional data fields

```python
additionalFieldsColumns = ['withGender', 'withBirthDate', 'withDeathDate']
utils_plot.plotPercentageTimeline(personCorrelations,
                                  additionalFieldsColumns,
                                  'Person correlation list - additional data',
                                 'numberEntries')
```

```python
personCorrelations[['numberEntries'] + additionalFieldsColumns]
```

---
### Multiple data source identifiers (1:n)

```python
multipleSourceIdentifierColumns = ['withMultipleKBR', 'withMultipleBnF', 
                                   'withMultipleNTA', 'withMultipleUnesco',
                                   'withMultipleVIAF', 'withMultipleISNI', 
                                   'withMultipleWikidata']
utils_plot.plotPercentageTimeline(personCorrelations,
                                  multipleSourceIdentifierColumns,
                                  'Person correlation list - 1:n identifiers',
                                 'numberEntries')
```

```python
personCorrelations[['numberEntries'] + multipleSourceIdentifierColumns]
```

```python
duplicateSourceIdentifierColumns = ['duplicateKBR', 'duplicateBnF', 
                                   'duplicateNTA', 'duplicateUnesco',
                                   'duplicateVIAF', 'duplicateISNI', 
                                   'duplicateWikidata']
utils_plot.plotPercentageTimeline(personCorrelations,
                                  duplicateSourceIdentifierColumns,
                                  'Person correlation list - n:1 identifiers',
                                 'numberEntries')
```

```python
personCorrelations[['numberEntries'] + duplicateSourceIdentifierColumns]
```
